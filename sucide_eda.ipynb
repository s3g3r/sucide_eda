```python
country_sucides = data1.groupby('country').sum()
```

```python
country_sucides.drop('year',axis=1,inplace=True)
```

```python
country_sucides.head()
```

```python
lowest_sucide = country_sucides.sort_values(by=['suicides/100k'])[:10]
lowest_sucide.head(5)
```

```python
highest_sucide = country_sucides.sort_values(by=['suicides/100k'],ascending=False)[:10]
highest_sucide.head(5)
```

```python
highest_sucide.reset_index(inplace=True)
lowest_sucide.reset_index(inplace=True)
```

```python
plt.figure(figsize=(12,8))

sns.barplot(data=lowest_sucide,x='country',y='suicides/100k')
plt.xticks(Rotation=70)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=highest_sucide,x='country',y='suicides/100k')
plt.xticks(Rotation=70)
```

```python
grp = data[['country','year','sex','population','suicides/100k pop']]
grp.head(3)
```

```python
ff = grp.groupby('country').sum()
ff.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,15))
sns.barplot(y='country',x='population',data=ff.sort_values(by=['population'], ascending=False)[:20])
plt.xlabel('Population Count')
plt.ylabel('Countries')
plt.title('2016 Year Sum Population for Suicide Rate')
plt.show()
```
